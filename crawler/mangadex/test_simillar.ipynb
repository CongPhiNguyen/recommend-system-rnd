```python
def intersectLength(arr1, arr2):
  m = {}
  if len(arr1)<len(arr2):
      arr1,arr2 = arr2,arr1
  for i in arr1:
      if i not in m:
        m[i] = 1
      else:
        m[i]+=1
  result = []
  for i in arr2:
      if i in m and m[i]:
        m[i]-=1
        result.append(i)
  return len(result)
def computeSimilar(dictVal0, dictVal1):
  tags0 = dictVal0["tags"][0].split(",")
  tags1 = dictVal1["tags"][0].split(",")
  return intersectLength(tags1, tags0)


matrix = []
for i in range(0,313):
  matrixTempt = []
  for j in range(0,313):
    matrixTempt.append(computeSimilar(dictVal0=dictVal[i], dictVal1=dictVal[j]))
  matrix.append(matrixTempt)

print(matrix)
```
